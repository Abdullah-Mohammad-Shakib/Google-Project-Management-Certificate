```python
import re

def remove_html_tags(text):
    clean = re.compile('<.*?>')
    return re.sub(clean, '', text)
```

```python
import re

def remove_urls(text):
    url_pattern = r'https?://\S+|www\.\S+'
    return re.sub(url_pattern, '', text)

# Example usage
text_with_urls = "Check out this link: https://example.com and also visit www.example.org for more info."
clean_text = remove_urls(text_with_urls)
print(clean_text)
```

This is the example of code that how we can remove URL from the text.

```python
exclude = string.punctuation
```

```python
def remove_punc(text):
    for char in exclude:
        text= text.replace(char , ' ')
        return text
```

```python
text='Hi! I who are you? can i help you?'
```

```python
start = time.time()
print(remove_punc(text))
time1 = time.time()- start
print(time1*50000)
```
